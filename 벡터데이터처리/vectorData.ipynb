```python
# 세계 지도 데이터 불러오기(GeoPandas 내장 지도)
# world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
import requests
import os
import zipfile
import io

def download_natural_earth_data():
    # URL of the Natural Earth data
    url = "https://naciscdn.org/naturalearth/110m/cultural/ne_110m_admin_0_countries.zip"
    
    # Download the zip file
    response = requests.get(url)
    
    # Create data directory if it doesn't exist
    if not os.path.exists('data'):
        os.makedirs('data')
    
    # Extract the contents
    z = zipfile.ZipFile(io.BytesIO(response.content))
    z.extractall('data')
    
    # Read the shapefile
    world = gpd.read_file('data/ne_110m_admin_0_countries.shp')
    return world

# 데이터 다운로드 및 로드
data = download_natural_earth_data()
print(list(data.columns))
```

```python
world = data.loc[:, ['POP_EST', 'CONTINENT', 'NAME', 'ISO_A3', 'geometry']]
world.head(1)
```

```python

world.head()
```

```python
world.plot()
```

```python
# 세계 지도(world)에서 대륙(continent) 열을 사용하여 아시아 국가만을 추출
asia = world[world['CONTINENT'] == 'Asia']
asia.plot()
```

```python
import requests
import os
import zipfile
import io

#세계 주요 도시 데이터 불러오기
# cities = gpd.read_file(gpd.datasets.get_path('naturalearth_cities'))
def download_cities_data():
    # URL for populated places data
    url = "https://naciscdn.org/naturalearth/110m/cultural/ne_110m_populated_places.zip"
    
    # Download the data
    response = requests.get(url)
    
    # Create data directory if it doesn't exist
    if not os.path.exists('data'):
        os.makedirs('data')
    
    # Extract the contents
    z = zipfile.ZipFile(io.BytesIO(response.content))
    z.extractall('data')
    
    # Read the shapefile
    cities = gpd.read_file('data/ne_110m_populated_places.shp')
    return cities

# 데이터 다운로드 및 로드
cities = download_cities_data()
cities
```

```python
cities = cities.loc[:, ['NAME', 'geometry']]
cities
```

```python
# 세계 주요 도시에서 아시아 도시만을 추출
asia_cities = gpd.overlay(cities, asia, how = 'intersection')

# 결과 확인
asia_cities
```

```python
# 결과 GeoDataFrame을 shapefile로 저장
asia_cities.to_file('./asia_cities.shp')
```

```python
# 유럽 대륙 추출
europe = world[world['CONTINENT'] == 'Europe']

# 유럽 + 아시아
eurasia = asia.overlay(europe, how = 'union', keep_geom_type=False)
```

```python
eurasia.plot()
```

```python
# to_crs 함수를 사용하여 좌표계 변환
asia = asia.to_crs(5179)
asia_cities = asia_cities.to_crs(5179)

# 버퍼 GeoDataFrame 생성을 위해 복사본 생성
asia_cities_buffer = asia_cities.copy()
asia_cities_buffer
```

```python
# asia_cities_buffer의 geometry열에 buffer 함수를 사용하여 갱신
asia_cities_buffer['geometry'] = asia_cities.buffer(100000)
asia_cities_buffer
```

```python
# 아시아 대륙과 아시아 주요 도시 버퍼의 대칭 차집합
sd_asia = gpd.overlay(asia, asia_cities_buffer, how = 'symmetric_difference', keep_geom_type=False)
sd_asia.plot()
```

```python
# 시군구 shp 파일 불러오기
gdf = gpd.read_file('./vectorData/sigungu.shp', encoding='cp949')
gdf.to_crs(5179)
gdf.head()
```

```python
# 법정동코드(BJCD)로부터 시도코드 추출
gdf['SDCD'] = gdf['BJCD'].str[:2]
gdf.head()
```

```python
gdf.plot(edgecolor = "black")
```

```python
# SDCD(시도코드)를 기준으로 dissolve 실행
dissolve_result = gdf.dissolve(by = 'SDCD')

dissolve_result.plot(edgecolor = "black")
```
