# Raster Data
주요 기본 개념
- 래스터 데이터 : 격자 형태의 공간 데이터
- 밴드 : 래스터 데이터의 층
- 해상도 : 격자의 크기
- 좌표계 : 공간 참조 시스템
- 모자이크 : 여러 래스터 데이터를 하나로 병합

## 래스터 데이터와 벡터 데이터 비교
- 래스터 데이터
   + 격자(grid) 형태로 공간을 표현
   + 연속적인 현상 표현에 적합 (고도, 기온, 강수량 등)
   + 격자 크기에 따라 해상도가 결정됨
   + 용량이 큰 편이며 격자 크기가 작을수록 용량 증가

- 벡터 데이터  
   + 점, 선, 면으로 공간을 표현
   + 이산적인 현상 표현에 적합 (도로, 건물, 행정구역 등)
   + 정확한 위치 표현 가능
   + 상대적으로 용량이 작음


## 레스터 구조의 특성
| 구분 | 설명 | 예시 |
|------|------|------|
| 셀 값 | - 각 셀은 하나의 숫자값을 가짐 | - 고도: 100m |
|       | - 숫자값은 해당 위치의 속성을 나타냄 | - 토지피복: 1(시가화지역), 2(농업지역) |
| 셀 크기 | - 모든 셀의 크기가 동일함 | - Landsat: 30m x 30m |
|          | - 셀 크기가 작을수록 해상도가 높아짐 | - KOMPSAT: 1m x 1m |
| 밴드 | - 하나의 래스터는 여러 개의 밴드로 구성될 수 있음 | - RGB 영상: 3개 밴드 |
|      | - 각 밴드는 서로 다른 정보를 담고 있음 | - Landsat: 다중분광 7개 밴드 |
| 공간 참조 | - 좌표체계 정보를 포함 | - WGS84 |
|          | - 지리좌표계 또는 투영좌표계 사용 | - EPSG:5174(중부원점) |

## 래스터 데이터의 종류
- USGS(United States Geological Survey, 미국지질조사국)
     + 위성영상, 항공사진, 지형도 등 다양한 공간정보 제공
     + Landsat, Sentinel 등 위성영상 무료 제공
- EarthExplorer
     + USGS에서 운영하는 위성영상 다운로드 플랫폼
     + 회원가입 후 무료로 위성영상 다운로드 가능
     + 다양한 위성영상(Landsat, Sentinel 등) 검색 및 다운로드
- 수치표고모델(DEM, Digital Elevation Model)
     + 지표면의 고도값을 격자 형태로 표현한 데이터
     + 국토지리정보원 홈페이지에서 제공
     + 해상도: 1m, 5m, 10m, 30m, 90m 등
     + 지형분석, 수문분석 등에 활용
- 토지피복지도
     + 지표면의 피복 상태를 분류한 주제도
     + 환경부 환경공간정보서비스(https://egis.me.go.kr)에서 제공
     + 대분류(7개 항목), 중분류(22개 항목), 세분류(41개 항목)
     + 도시계획, 환경영향평가 등에 활용

```python
# 입력 파일 경로 지정
raster_files = ['./rasterData/토지피복도1.tif', './rasterData/토지피복도2.tif']

# 결과 파일을 담을 임시 리스트 생성
raster_to_merge = []

# 결과 파일 이름 및 경로 지정
output_path = './rasterData/merge_result.tif'
```

```python
# 반복문을 사용하여 파일 불러오기(2개 영상)
for i in raster_files:
    raster = rio.open(i)
    raster_to_merge.append(raster)
```

```python
# 1번 영상 확인
show(raster_to_merge[0])
```

```python
# 2번 영상 확인
show(raster_to_merge[1])
```

```python
# 래스터 파일 합치기(merge)
merge, output = merge(raster_to_merge)
```

```python
# 메타데이터를 복사하여 신규 변수에 할당
output_meta = raster.meta.copy()

# 복사된 메타데이터 정보를 갱신
output_meta.update(
    {"driver": "GTiff",               # 저장 포맷 설정
        "height": merge.shape[1],  # 높이 설정
        "width": merge.shape[2],   # 너비 설정
        "transform": output}        # 변환 옵션 지정
)
```

```python
# 파일 내보내기(저장하기)
# (이전 코드) output_path = ./merge_result.tif

with rio.open(output_path, "w", **output_meta) as m:
    m.write(merge)
```

```python
# mask(clip) 기능 사용을 위한 패키지 불러오기
from rasterio.mask import mask
import fiona

# 입력 데이터 불러오기
input_image = "./rasterData/merge_result.tif"
```

```python
# 거푸집 Shapefile로부터 자를 영역의 정보 추출
with fiona.open("./rasterData/mask_rec.shp", "r") as shapefile:
    features = [feature["geometry"] for feature in shapefile]
```

```python
# mask 작업 실행 및 메타데이터 복사본 생성
with rio.open(input_image) as ori_image:
    out_image, out_transform = mask(ori_image, features, crop = True)
    out_meta = ori_image.meta.copy()
```

```python
out_image
```

```python
# 메타데이터 정보 갱신
out_meta.update({
    "driver": "GTiff",
    "height": out_image.shape[1],
    "width": out_image.shape[2],
    "transform": out_transform
})
```

```python
# 결과 파일 저장
with rio.open('./rasterData/result_mask.tif', "w", **out_meta) as dest:
    dest.write(out_image)
```
